### Import modules

```python
from dataset.language_dataset.language.totto.baseline_preprocessing.preprocess_data_main import _generate_processed_examples
from dataset.language_dataset.language.totto.baseline_preprocessing import preprocess_utils
import pandas as pd
```

### Import dataset

```python
preprocessed_samples=_generate_processed_examples("dataset/totto_dataset/totto_train_data.jsonl")
```

### Converting train data to csv file

```python
data={"table":[], "sentence":[]}
for table in preprocessed_samples:
    data['table'].append(table['subtable_metadata_str'])
    data['sentence'].append(table['sentence_annotations'][0]['final_sentence'])
```

```python
table_csv = pd.DataFrame(data)
table_csv.to_csv("Preprocessed_Train_Dataset.csv")
```
